# 반복문
### 반복문을 이용한 데이터프레임 접근

- 반복문은 기존 Python 문법에서 배운 내용
- 반복문이 필요한 이유는? 데이터프레임 전처리 과정에서 다양한 데이터를 빠르게 전처리 하기 위함

```python
#흔히 말하는 for문을 복습
#range 함수를 이용한 반복문!
for i in range(10):
    print(i+100)
```

```python
#시퀀스객체인 리스트를 이용한 반복문 진행해 보자!
a= ['파문기','파문응','데분기','데분중','데분고','SQL']
```

```python
#시퀀스객체 그대로 가지고 오기 
for i in a:
    print(i)
```

```python
#인덱스로 가지고 오기
for i in range(len(a)):
    print(a[i])
```

- 반복문 중 값과 인덱스를 같이 보는 방법은?

```python
#enumerate 처음은 인덱스, 뒤에는 값으로
for i,j in enumerate(a):
    print(a[i],j)
```

```python
#zip 함수 이용해서도 가지고 올 수 있다.
b = [100,200,300,400,500,600]
```

```python
# i는 a에 대한 루프인덱스, j는 b에대한 루프인덱스
for i,j in zip(a,b):
    print(i,j)
```

```python
#응용도 가능하다!
for i , (j ,k) in enumerate(zip(a, b)):
    print(i,j,k)
```

```python
import itertools
#라이브러리 중 반복문 같이 가지고 올 수 있는 중첩을 하지 않아도 두 개를 같이 가지고 올 수 있다.
for i,j in itertools.product(a,b):
    print(i,j)
```

### 데이터프레임으로 반복문을 접근할 때 중요한 점은!

### 반복문의 input output에 대한 원리를 이해하면 쉽게 사용할 수 있다!

```python
#데이터프레임을 그냥 바로 넣어서 반복문을 실행하면
for i in df:
    print(i)
# 데이터프레임을 그대로 넣게 되면 바로 루프인덱스는 컬럼을 출력한다.
```

```python
#데이터프레임을 그냥 바로 넣어서 반복문을 실행하면
for i in df.model:
    print(i)
```

```python
#데이터프레임을 그냥 바로 넣어서 반복문을 실행하면
for i in df.cty:
    print(i)
```

```python
##응용을 해보자!
for i,j in zip(df.cty, df.hwy):
    print(i,j)
```

```python
##이 두개의 연비 평균을 내면, 차량의 평균 연비가 나온다.
for i,j in zip(df.cty, df.hwy):
    print((i+j)/2)
```

```python
## 해당 값을 데이터프레임에 넣어보자! df에 새로운 파생변수를 만들어보자
## 리스트에 값을 넣는법
# append이용해서 리스트에 값을 넣으려면?
mc = []
for i,j in zip(df.cty, df.hwy):
    mc.append((i+j)/2)
    #print((i+j)/2)
```

```python
#데이터 프레임에 값을 넣는법?
#데이터프레임에 새로운 파생변수
df['mc'] =mc
```
